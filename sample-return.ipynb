# ETAPAS DO PROCESSO DE ESTUDO

## 1 - ESCOLHER O DATA SET DE ESTUDO
    PARA ESSE PROJETO, DATA SET PROPOSTO FOI UM CONJUNTO DE DADOS COM AMOSTRAS DAS SEIS MISSÕES APOLLO QUE ATERRISARAM NA LUA.

    APÓS BAIXAR OS DADOS INICIAMOS O LEITURA DOS DADOS UTILIZANDO PANDAS.

```python
import pandas as pd

rock_samples = pd.read_csv("rocksamples.csv")
```

```python
rock_samples.head()
```

Para saber obter mais informações sobre a amostra, podemos rodar o comando .info().

Como podemos ver abaixo, a mostra possui um conjunto de 2229 linhas; por seis colunas;


```python
rock_samples.info()
```

Para faciliar a compreensão dos dados, renomeei as colunas.

Afim de não perder a integrade dos dados brutos, primeiro seto o dataset em um dataframe no pandas

```python
df = rock_samples

df.rename(columns={'Mission': 'Missao'}, inplace = True)
df.rename(columns={'Type': 'Tipo'}, inplace = True)
df.rename(columns={'Subtype': 'Subtipo'}, inplace = True)
df.rename(columns={'Weight (g)': 'Peso (g)'}, inplace = True)
df.rename(columns={'Pristine (%)': 'Integridade (%)'}, inplace = True)

df.head()
```

## 2 - CONVERTER UNIDADE DE PESO DA AMOSTRA

 DEVIDO AO CONTEXTO, LEVAMOS EM CONSIDERAÇÃO QUE O PESO DO FOGUETE É MEDIDO EM QUILOGRAMAS.
 ESSA É UMA CONVERSÃO BEM SIMPLES, UTILIZANDO EXPRESSÃO LAMBDA, OU PSEUDO FUNÇÃO, PODEMOS DIZER QUE UM GRAMA EQUIVALE A 0,001 KG

```python
df['Peso (g)'] = df['Peso (g)'].apply(lambda x :x * 0.001)
df.rename(columns={'Peso (g)': 'Peso (kg)'}, inplace = True)
df.head()
```

## 3 - DEFINIR DATAFRAME DE ANALISE

ESSE DATA FRAME É BASICAMENTE UM RESUMO DA AMOSTRA. ELE IRÁ CONTER OS DADOS DE CADA MISSÃO CONTIDA NA AMOSTRA

```python
missoes = pd.DataFrame()
missoes['Missao'] = df['Missao'].unique()
missoes.head()
```

```python
missoes.info()
```

Iniciamos os totatilzadores (importante manter ter cuidado com a nomenclatura chave das colunas)

```python
sample_total_weight = df.groupby('Missao')['Peso (kg)'].sum()
missoes = pd.merge(missoes, sample_total_weight, on='Missao')
missoes.rename(columns={'Peso (kg)':'Peso amostra (kg)'}, inplace=True)
missoes
```

Se você fosse imprimir essa linha, teria uma série pandas, que é basicamente um tipo de dados em 1D, ou uma lista. A lista teria como índice o valor exclusivo da coluna Missão em vez de um número:

```python
sample_total_weight = df.groupby('Missao')['Peso (kg)'].sum()
sample_total_weight
```

```python
missoes['DiferencaPeso'] = missoes['Peso amostra (kg)'].diff()
missoes
```

Substituimos valores NaN (Not is a Number) para 0

```python
missoes['DiferencaPeso'] = missoes['DiferencaPeso'].fillna(value=0)
missoes
```

# 4 - ADICIONAMOS AO DATASET DADOS DE MODULOS LUNARES E DE COMANDO

```python
missoes['Modulo lunar (LM)'] = 'Eagle (LM-5)', 'Intrepid (LM-6)', 'Antares (LM-8)', 'Falcon (LM-10)', 'Orion (LM-11)', 'Challenger (LM-12)'
missoes['LM massa (kg)'] = 15103, 15235, 15264, 16430, 16445, 16456
missoes['LM massa diff'] = missoes['LM massa (kg)'].diff()
missoes['LM massa diff'] = missoes['LM massa diff'].fillna(value=0)

missoes['Modulo comando (CM)'] = 'Columbia (CSM-107)', 'Yankee Clipper (CM-108)', 'Kitty Hawk (CM-110)', 'Endeavor (CM-112)', 'Casper (CM-113)', 'America (CM-114)'
missoes['CM massa (kg)'] = 5560, 5609, 5758, 5875, 5840, 5960
missoes['CM massa diff'] = missoes['CM massa (kg)'].diff()
missoes['CM massa diff'] = missoes['CM massa diff'].fillna(value=0)

missoes
```

Adicionamos totalizadores para cada missao

```python
missoes['Total peso (kg)'] = missoes['LM massa (kg)'] + missoes['CM massa (kg)']
missoes['Total peso dif'] = missoes['LM massa diff'] + missoes['CM massa diff']
missoes
```

```python
# Sample-to-weight ratio
saturnVPayload = 43500
missoes['Crewed area : Payload'] = missoes['Total peso (kg)'] / saturnVPayload
missoes['Sample : Crewed area'] = missoes['Peso amostra (kg)'] / missoes['Total peso (kg)']
missoes['Sample : Payload'] = missoes['Peso amostra (kg)'] / saturnVPayload
missoes
```

```python
crewedArea_payload_ratio = missoes['Crewed area : Payload'].mean()
sample_crewedArea_ratio = missoes['Sample : Crewed area'].mean()
sample_payload_ratio = missoes['Sample : Payload'].mean()
print(crewedArea_payload_ratio)
print(sample_crewedArea_ratio)
print(sample_payload_ratio)
```
